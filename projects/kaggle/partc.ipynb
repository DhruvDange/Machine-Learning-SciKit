```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dxdt (x, omega, t):
  return (gamma*x*(1-x) - h*(1+(eta*np.cos(omega*t))))

gamma = 5.0
h = 1
eta = 0.5
x0 = 0.4
#omega = 2*3.14
ti = 0 
tf = 5
dt = 0.01
```

```python
def euler_solve (omega, dt):
  x=np.zeros(len(t))
  x[0] = x0
  for i in range(0, len(t)-1):
    if(x[i] <0 ):
      x[i] = 0
      break
    x[i+1]=x[i] + dt* dxdt(x[i],omega,i)
  return x
```

```python
omega = np.arange(0,2*np.pi,0.01)
t= np.arange(ti,tf,dt)

for i in range(0,len(omega)):
    xsol = euler_solve(omega[i], dt)
    plt.rcParams["figure.figsize"] = (16, 9)
    plt.plot(t,xsol)
    
plt.show()
```
