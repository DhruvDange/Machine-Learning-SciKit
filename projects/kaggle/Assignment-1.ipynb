```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def dxdt (x, t):
  return (gamma*x*(1-x) - h*(1+(eta*np.cos(omega*t))))

gamma = 5.0
h = 1.2006
eta = 0.5
omega = 2*3.14
ti = 0 
tf = 6
dt = 0.001
```

```python
def euler_solve (x0, dt):
  x=np.zeros(len(t))
  x[0] = x0
  for i in range(0, len(t)-1):
    if(x[i] <0 ):
      break
    x[i+1]=x[i] + dt* dxdt(x[i],i)
  return x
```

```python
x = np.arange(0,1,0.01)
t= np.arange(ti,tf,dt)

for i in range(0,len(x)):
    xsol = euler_solve(x[i], dt)
    plt.rcParams["figure.figsize"] = (16, 9)
    plt.plot(t,xsol)
    
plt.show()
```
