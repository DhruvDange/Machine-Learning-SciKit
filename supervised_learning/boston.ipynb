```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import mglearn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
print("Training score: {:.2f}".format(linear_reg.score(X_train, y_train)))
print("Test score: {:.2f}".format(linear_reg.score(X_test, y_test)))
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_obj10 = Ridge(alpha=10).fit(X_train, y_train)
print("Training score: {:.2f}".format(ridge_obj10.score(X_train, y_train)))
print("Test score: {:.2f}".format(ridge_obj10.score(X_test, y_test)))
```

```python
ridge_obj01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Training score: {:.2f}".format(ridge_obj01.score(X_train, y_train)))
print("Test score: {:.2f}".format(ridge_obj01.score(X_test, y_test)))
```

```python
ridge_obj = Ridge(alpha=1).fit(X_train, y_train)
print("Training score: {:.2f}".format(ridge_obj.score(X_train, y_train)))
print("Test score: {:.2f}".format(ridge_obj.score(X_test, y_test)))
```

```python
plt.plot(ridge_obj.coef_, 's', label="Ridge alpha=1")
plt.plot(ridge_obj10.coef_, '^', label="Ridge alpha=10")
plt.plot(ridge_obj01.coef_, 'v', label="Ridge alpha=0.1")
plt.plot(linear_reg.coef_, 'o', label="LinearRegression")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.hlines(0, 0, len(linear_reg.coef_))
plt.ylim(-25, 25)
plt.legend()
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso.score(X_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso.coef_ != 0)))
```

```python
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso001.score(X_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso001.coef_ != 0)))
```

```python
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso00001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso00001.score(X_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso00001.coef_ != 0)))
```

```python
plt.plot(lasso.coef_, 's', label="Lasso alpha=1")
plt.plot(lasso001.coef_, '^', label="Lasso alpha=0.01")
plt.plot(lasso00001.coef_, 'v', label="Lasso alpha=0.0001")
plt.plot(ridge_obj01.coef_, 'o', label="Ridge alpha=0.1")
plt.legend(ncol=2, loc=(0, 1.05))
plt.ylim(-25, 25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
```
